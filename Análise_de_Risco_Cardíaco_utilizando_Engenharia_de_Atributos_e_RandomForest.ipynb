Coment√°rios explicativos e organiza√ß√£o em se√ß√µes claras.

```python
#===============================================
# 1. Instalar e importar reposit√≥rio do dataset
#===============================================
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

#=========================
# 2. Importar bibliotecas
#=========================
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, roc_auc_score
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from numpy import mean # Para calcular a m√©dia
import matplotlib.pyplot as plt
import seaborn as sns

#========================
# 3. Carregar o dataset
#========================
heart_disease = fetch_ucirepo(id=45) # importa dataset direto, sem arquivo csv
X = heart_disease.data.features
y = heart_disease.data.targets

#====================================================
# 4. Pr√©-processamento e Engenharia de Atributos (FE)
#====================================================

# 4.1. Binariza√ß√£o do Target y
# 0 = Aus√™ncia; 1, 2, 3, 4 = Presen√ßa. Binarizamos para 0 ou 1.
y_binarized = (y > 0).astype(int)
y_binarized.columns = ['target']

# 4.2. Tratamento de NaNs (Imputa√ß√£o simples com a moda/valor mais frequente)
# Isso √© mais seguro do que remover linhas (dropna)
# Colunas com NaNs s√£o 'ca' e 'thal'
for col in ['ca', 'thal']:
    if col in X.columns:
        moda = X[col].mode()[0] # Calcula a moda (o valor mais frequente)
        X[col].fillna(moda, inplace=True)

# 4.3. Engenharia de Atributos (3 Novas Vari√°veis)
# A cria√ß√£o de novas vari√°veis (FE) vem antes do get_dummies, para no fim todas terem dados num√©ricos
X_fe = X.copy()

# 1. stress_index: Combina√ß√£o de oldpeak e exang (depress√£o do segmento ST e angina - dor no peito)
X_fe['stress_index'] = X_fe['oldpeak'] * (1 + X_fe['exang'])

# 2. chol_trestbps_ratio: Raz√£o entre colesterol e press√£o arterial
X_fe['chol_trestbps_ratio'] = X_fe['chol'] / (X_fe['trestbps'] + 1)

# 3. age_ca_interaction: Intera√ß√£o entre a idade e o n√∫mero de vasos afetados
X_fe['age_ca_interaction'] = X_fe['age'] * X_fe['ca']


# 4.4. One-Hot Encoding (OHE) para Categ√≥ricas
# cp, restecg, slope, thal s√£o categ√≥ricas/ordinais
X_final = pd.get_dummies(X_fe, columns=['cp', 'restecg', 'slope', 'thal'], drop_first=True)

print(f"\nN√∫mero final de atributos: {X_final.shape[1]}")

#========================================
# 5. Dividir o dataset em treino e teste
#========================================
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_binarized['target'], test_size=0.2, random_state=42, stratify=y_binarized # 20% teste /80% treino
)

#=================================================
# 6. Criar Modelo de Floresta Aleat√≥ria (3 Varia√ß√µes)
#=================================================
# VARIAC√ÉO 1: Modelo A (Baseline: Par√¢metros Default)
print("\n--- VARIAC√ÉO 1: MODELO BASELINE (Default) ---")
rf_baseline = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=None, min_samples_leaf=1)
rf_baseline.fit(X_train, y_train)

y_train_pred_A = rf_baseline.predict(X_train)
y_test_pred_A = rf_baseline.predict(X_test)

print(f"Acur√°cia TREINO (A): {accuracy_score(y_train, y_train_pred_A):.4f}")
print(f"Acur√°cia TESTE (A):  {accuracy_score(y_test, y_test_pred_A):.4f}")
print("Relat√≥rio de Classifica√ß√£o TESTE (A):")
print(classification_report(y_test, y_test_pred_A))


# VARIAC√ÉO 2: Modelo B (Overfitting/Alto Risco)
print("\n--- VARIAC√ÉO 2: MODELO DE ALTO RISCO (max_depth alto) ---")
# Usamos alta profundidade e n√∫mero fixo de estimadores para demonstrar sobreajuste
rf_overfit = RandomForestClassifier(n_estimators=300, random_state=42, max_depth=30, min_samples_leaf=1)
rf_overfit.fit(X_train, y_train)

y_train_pred_B = rf_overfit.predict(X_train)
y_test_pred_B = rf_overfit.predict(X_test)

print(f"Acur√°cia TREINO (B): {accuracy_score(y_train, y_train_pred_B):.4f}")
print(f"Acur√°cia TESTE (B):  {accuracy_score(y_test, y_test_pred_B):.4f}")
print("Relat√≥rio de Classifica√ß√£o TESTE (B):")
print(classification_report(y_test, y_test_pred_B))


# VARIAC√ÉO 3: Modelo C (Otimizado via GridSearchCV)
print("\n--- VARIAC√ÉO 3: MODELO OTIMIZADO (GridSearchCV) ---")
param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [10, 20, None],
    'min_samples_leaf': [1, 5]
}

rf_tuned = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf_tuned,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=0
)

grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

y_train_pred_C = best_model.predict(X_train)
y_test_pred_C = best_model.predict(X_test)

print(f"Melhores Par√¢metros encontrados: {best_params}")
print(f"Acur√°cia TREINO (C): {accuracy_score(y_train, y_train_pred_C):.4f}")
print(f"Acur√°cia TESTE (C):  {accuracy_score(y_test, y_test_pred_C):.4f}")
print("Relat√≥rio de Classifica√ß√£o TESTE (C):")
print(classification_report(y_test, y_test_pred_C))

#====================================================
# 7. Sele√ß√£o e An√°lise do melhor Modelo (Modelo C)
#====================================================

print("\n=========================================================")
print("RESULTADO OBTIDO COM O MODELO OTIMIZADO (COM FE E TUNING)")
print("===========================================================")

# Apresenta√ß√£o das m√©tricas FINAIS (Acur√°cia, Precision, Recall, F1, AUC)
y_proba_opt = best_model.predict_proba(X_test)[:, 1]

print("\n--- M√©tricas Finais do Melhor Modelo ---")
print(f"Acur√°cia: {accuracy_score(y_test, y_test_pred_C):.4f}")
print(f"AUC Score: {roc_auc_score(y_test, y_proba_opt):.4f}")
print("Relat√≥rio de Classifica√ß√£o (Precision, Recall, F1):")
print(classification_report(y_test, y_test_pred_C))


# 7.1. Gr√°fico de Import√¢ncia das Vari√°veis (Feature Importance)
importances = best_model.feature_importances_
feature_names = X_final.columns
feature_importance_df = pd.DataFrame({
    'Atributo': feature_names,
    'Import√¢ncia': importances
}).sort_values(by='Import√¢ncia', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Import√¢ncia', y='Atributo', data=feature_importance_df.head(10), palette='viridis')
plt.title('Top 10 Import√¢ncia de Vari√°veis (Melhor Modelo)')
plt.show()

# 7.2. Matriz de Confus√£o
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_C)
plt.title('Matriz de Confus√£o do Melhor Modelo')
plt.show()

```

##**EXPLICA√á√ÉO 2. Engenharia de atributos**

**Novo Atributo: stress_index**

**Hip√≥tese e contribui√ß√£o:** A combina√ß√£o do grau de depress√£o do segmento ST (oldpeak) com a presen√ßa de dor no peito induzida por exerc√≠cio (exang) √© um marcador mais forte de isquemia grave do que os fatores isolados. Contribui√ß√£o: Cria um √≠ndice composto que amplifica o sinal de risco quando ambos os fatores de estresse est√£o presentes.

**Novo Atributo: chol_trestbps_ratio**

**Hip√≥tese e contribui√ß√£o:** A propor√ß√£o entre o colesterol (chol) e a press√£o arterial de repouso (trestbps) reflete o risco metab√≥lico geral versus o risco de hipertens√£o. Contribui√ß√£o: Permite ao modelo capturar intera√ß√µes entre fatores cr√¥nicos (colesterol) e agudos (press√£o arterial), que podem ser mais preditivas do que os valores brutos.

**Novo Atributo: age_ca_interaction**

**Hip√≥tese e contribui√ß√£o:** O impacto do n√∫mero de vasos sangu√≠neos afetados (ca) √© mais perigoso √† medida que a idade (age) aumenta. Contribui√ß√£o: Introduz um termo de intera√ß√£o que penaliza mais os pacientes mais velhos com m√∫ltiplas obstru√ß√µes, algo que os modelos lineares teriam dificuldade em aprender.
